```python

from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Voy a usar el archivo tableParamWithFitness.csv obtenido tras finalizar una simulación de la estrategia 3
#Abri el archivo y me aseguro que el header y los datos están con el mismo formato
#Si no lo están, el archivo dará problemas a la hora de leer el csv con python

data = pd.read_csv("tableParamWithFitness_E3b.csv",sep=',',header=(0)) 
data.head()
```

```python
gly1=data["gly1"] #eje x
gly2=data["gly2"] #eje y
fitness=data["fitness"] #eje z
```

```python
gly1
```

```python
ax = plt.axes(projection='3d')
# Data for three-dimensional scattered points
ax.set_xlabel('gly1')
ax.set_ylabel('gly2')
ax.set_zlabel('fitness')

ax.scatter3D(gly1, gly2, fitness, c=fitness, cmap='Greens');
plt.savefig('Mapa_parametros_3D.png', dpi=700,transparent=True)

```

```python
#Fancy dic: https://stackoverflow.com/questions/18491093/is-there-a-way-to-have-a-dictionary-key-be-a-range
def fancy_dict(*args):
    'Pass in a list of tuples, which will be key/value pairs'
    ret = {}
    for k,v in args:
        for i in k:
            ret[i] = v
    return ret

colors = fancy_dict((range(0,3),"Red"))
```

```python
ndata=len(data.gly1)
fitnesclass= np.array(ndata*[["NA"]], dtype=object)
ndata=len(data.fitness)
for i in range(0,ndata):
    
    if data.fitness[i] <=1:
        fitnesclass[i]="0-1"
    elif 1<data.fitness[i]<=2:
        fitnesclass[i]="1-2"
    elif 2<data.fitness[i]<=3:
        fitnesclass[i]="2-3"
    elif 3<data.fitness[i]<=3.3:
        fitnesclass[i]="3-3.3"
    elif 3.3<data.fitness[i]<=3.35:
        fitnesclass[i]="3.3-3.35"
    elif 3.35<data.fitness[i]<=3.4:
        fitnesclass[i]="3.35-3.4"
    elif 3.40<data.fitness[i]<=3.5:
        fitnesclass[i]="3.4-3.5"
    elif data.fitness[i]>3.5:
        fitnesclass[i]=">3.5"


```

```python
color=pd.DataFrame(fitnesclass, columns=['Color']) 
color
```

```python
filtered_data = pd.DataFrame(data)
cols = [2,3]
filtered_data = filtered_data[filtered_data.columns[cols]]
filtered_data

data2 = pd.concat([filtered_data, color],axis=1)
data2
```

```python
data2.Color
```

```python
colors={"0-1":"red","1-2":"darkorange","2-3":"gold","3-3.3":"yellow","3.3-3.35":"greenyellow","3.35-3.4":"limegreen","3.4-3.5":"green",">3.5":"darkviolet"}
fitness_colors=data2.Color.map(colors)
```

```python
fig,ax = plt.subplots()
ax.scatter(gly1, gly2, color=fitness_colors)

#Nombres ejes
ax.set_xlabel('gly1')
ax.set_ylabel('gly2')

plt.grid()

#Leyenda
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors.values()]
legend = plt.legend(markers, colors.keys(), bbox_to_anchor=(1.04,1), loc="upper left", title="Fitness", numpoints=1)
legend
plt.savefig('Mapa_colores_parametros_2D.png', bbox_extra_artists=(legend,), bbox_inches='tight', dpi=700)
plt.show()

#Para que se guarde la leyende también:
#https://www.iteramos.com/pregunta/23109/movimiento-matplotlib-leyenda-fuera-del-eje-hace-que-se-corte-por-la-figura-del-cuadro-de
```
